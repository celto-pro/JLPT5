## Alphabet
To train on alphabets, use the function "jap_alphabet.alphabet_training(12)". We can change 12 by any value between 1 and 69. This is the number of element of alphabet to train on. We can also specify another parameter if we want to focus on a specific translation sens:
- jap_alphabet.alphabet_training(12, specificity)

    - specificity=2 => (hiragana => romanji + katakana);

## First 80 kanjis

To train on the first 80 kanji for JLPT5, use the function "specific_kanjis_JLPT5" or "kanjis_training_JLPT5".
Here is a description of the parameters:
- 10*training_size represents the number of questions. Care of the factor 10. Better not use more than the value 5.
- difficulty is an integer or the value 'max'. It limits the number of different questions. It has to be higher than training_size
- direction can be 0 (or "jap => fr") or 1 (or "fr => jap"). It decides if the translation is from japanese to french or french to japanese.
- weight ("off" or 0) or ("on" or 1): "off" means that scores/level aren't taken in consideration in the chose of questions, "on" means it's taken in consideration.

```python
jap_JPTL5_80kanji.specific_kanjis_JLPT5(training_size=2, difficulty='max', direction=0, weight='off')
```

```python
jap_JPTL5_80kanji.kanjis_training_JLPT5(training_size=2, difficulty='max', weight="off")
```

### Direct training with "specific_vocabulary_JLPT5()" or "vocabulary_training_JLPT5()"
We can here train on the vocabulary with or without weights. We can also train on a specific range of the vocabulary. Here is a description of the parameters:
- training_size, difficulty, direction and weights are the same as in the part II. Kanjis.
- training_position is an additionnal parameter that can focus on a specific range of the vocabulary list. It's a list of two elements. For instance: [47, 276]

```python
jap_JLPT5_vocabulary.specific_vocabulary_JLPT5(training_size=3,
                                               difficulty='max',
                                               direction=0,
                                               weight='off',
                                               training_position=None)
```
